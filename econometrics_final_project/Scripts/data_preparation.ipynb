```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib
```

```python
pd.options.display.float_format = "{:,.2f}".format
```

```python
#loading ppi data less ppi for heat oil
dlho = pd.read_csv('/Users/jayco/Desktop/EBGN590/Final-Project/Data/data_less_heat_oil.csv')
dlho.head()
```

```python
#cleaning dataset
dlho = dlho.rename(columns={'CUSR0000SAH21_NBD20240201_CUSR0000SETB_NBD20240201': 'energy_index', 
                            'PCU21212121_NBD20240201': 'ppi_coal',
                            'PCU2111112111111_NBD20240201' : 'ppi_crude',
                            'PCU2111112111113_NBD20240201' : 'ppi_nat_gas',
                            'PCU324110324110121_NBD20240201' : 'ppi_gasoline',
                            'WPU057303_NBD20240201' : 'ppi_diesel'})
dlho.head()
```

```python
#final df for ppis
dho = pd.read_csv('/Users/jayco/Desktop/EBGN590/Final-Project/Data/heat_oil.csv')
heat_oil = dho['index_val'].values
dlho['ppi_heat_oil'] = heat_oil
dlho['ppi_heat_oil']=(dlho['ppi_heat_oil'] / dlho['ppi_heat_oil'].iloc[-1]) * 100
ppi_df = dlho
ppi_df
```

```python
#renewable consumption control
renew_cons=pd.read_csv('/Users/jayco/Desktop/EBGN590/Final-Project/Data/renew_cons.csv')
renew_df = DataFrame(renew_cons['Value'].values)
renew_df=renew_df.rename(columns={0 : 'renewable_residential_consumption'})
renew_df
```

```python
dxy_df = pd.read_csv('/Users/jayco/Desktop/EBGN590/Final-Project/Data/dxy.csv')
dxy = pd.DataFrame(dxy_df['Close'])
dxy = dxy.rename(columns={'Close': 'dxy'})
dxy['dxy']=(dxy['dxy'] / dxy['dxy'].iloc[-1]) * 100
dxy
```

```python
fedfunds_df = pd.read_csv('/Users/jayco/Desktop/EBGN590/Final-Project/Data/FEDFUNDS.csv')
fedfunds_df = fedfunds_df.rename(columns={'FEDFUNDS_PCH' : 'effr_change'})
fedfunds_df
```

```python
inflation_df = pd.read_csv('/Users/jayco/Desktop/EBGN590/Final-Project/Data/cpi.csv')
inflation_df = inflation_df.rename(columns={'1-Month % Change' : 'cpi_change'})
inflation = pd.DataFrame(inflation_df['cpi_change'])
inflation = inflation.iloc[:-1]
inflation
```

```python
dfs = [ppi_df, fedfunds_df, inflation, dxy, renew_df]

final_df = pd.concat(dfs,axis=1)
final_df = final_df.drop(final_df.columns[8],axis=1)
final_df.insert(0, 'DATE', ppi_df['DATE'])

final_df.head()
```

```python
final_df.to_csv('/Users/jayco/Desktop/EBGN590/Final-Project/Data/project_data.csv')
```
